```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100) 
```

```python
def exp_func():
    explist = []
    for i in x:
       explist.append(np.exp(i))
    return explist
```

```python
y = exp_func()
```

```python
plt.plot(x, y)
plt.xlabel = "Time [milliseconds]"
plt.ylabel = "Awesomeness"
plt.savefig(fname="Figure for CJ2p7", format="pdf")
```
