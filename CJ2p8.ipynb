```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sinx():
    sinxlist = []
    for i in x:
        sinxlist.append(np.sin(i))
    return sinxlist
```

```python
def cosx():
    cosxlist = []
    for i in x:
        cosxlist.append(np.cos(i))
    return cosxlist
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(sinx(), x)
ax[1].plot(cosx(), x)
ax[0].set_xlabel("sin(x)")
ax[1].set_xlabel("cos(x)")
fig.savefig(fname="CJ2p8 Figure", format="pdf")
```
